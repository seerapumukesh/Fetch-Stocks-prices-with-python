```python
import requests

def fetch_stock_news(api_key, symbols):
    base_url = "https://newsapi.org/v2/everything"
    news_articles = []

    for symbol in symbols:
        params = {
            'q': f'{symbol} stock',
            'apiKey': api_key
        }

        try:
            response = requests.get(base_url, params=params)
            response.raise_for_status()  # Raise an HTTPError for bad requests
            news_data = response.json()

            if news_data['status'] == 'ok':
                articles = news_data['articles']
                for article in articles:
                    news_articles.append({
                        'symbol': symbol,
                        'title': article['title'],
                        'author': article['author'],
                        'description': article['description'],
                        'url': article['url']
                    })
            else:
                print(f"Error for {symbol}: {news_data['message']}")

        except requests.exceptions.HTTPError as errh:
            print(f"HTTP Error for {symbol}:", errh)
        except requests.exceptions.ConnectionError as errc:
            print(f"Error Connecting for {symbol}:", errc)
        except requests.exceptions.Timeout as errt:
            print(f"Timeout Error for {symbol}:", errt)
        except requests.exceptions.RequestException as err:
            print(f"Oops! Something went wrong for {symbol}:", err)

    # Display the collected news articles
    for index, article in enumerate(news_articles, start=1):
        print(f"Article {index} ({article['symbol']}):")
        print(f"Title: {article['title']}")
        print(f"Author: {article['author']}")
        print(f"Description: {article['description']}")
        print(f"URL: {article['url']}")
        print("\n" + "="*50 + "\n")

if __name__ == "__main__":
    # Replace 'YOUR_API_KEY' with your News API key
    api_key = 'cf142db758e548d6ab60d5a51e61b208'

    # Replace the list with the stock symbols you're interested in
    symbols = ['WEN']

    fetch_stock_news(api_key, symbols)
```

```python
import requests
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def get_sentiment(text):
    analyzer = SentimentIntensityAnalyzer()
    compound_score = analyzer.polarity_scores(text)['compound']

    # Classify sentiment based on the compound score
    if compound_score >= 0.05:
        return 'Positive'
    elif compound_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

def fetch_stock_news(api_key, symbols):
    base_url = "https://newsapi.org/v2/everything"
    news_articles = []

    for symbol in symbols:
        params = {
            'q': f'{symbol} stock',
            'apiKey': api_key
        }

        try:
            response = requests.get(base_url, params=params)
            response.raise_for_status()  # Raise an HTTPError for bad requests
            news_data = response.json()

            if news_data['status'] == 'ok':
                articles = news_data['articles']
                for article in articles:
                    sentiment = get_sentiment(article['title'])
                    news_articles.append({
                        'symbol': symbol,
                        'title': article['title'],
                        'author': article['author'],
                        'description': article['description'],
                        'url': article['url'],
                        'sentiment': sentiment
                    })
            else:
                print(f"Error for {symbol}: {news_data['message']}")

        except requests.exceptions.HTTPError as errh:
            print(f"HTTP Error for {symbol}:", errh)
        except requests.exceptions.ConnectionError as errc:
            print(f"Error Connecting for {symbol}:", errc)
        except requests.exceptions.Timeout as errt:
            print(f"Timeout Error for {symbol}:", errt)
        except requests.exceptions.RequestException as err:
            print(f"Oops! Something went wrong for {symbol}:", err)

    # Display the collected news articles with sentiment
    for index, article in enumerate(news_articles, start=1):
        print(f"Article {index} ({article['symbol']}):")
        print(f"Title: {article['title']}")
        print(f"Author: {article['author']}")
        print(f"Description: {article['description']}")
        print(f"URL: {article['url']}")
        print(f"Sentiment: {article['sentiment']}")
        print("\n" + "="*50 + "\n")

if __name__ == "__main__":
    # Replace 'YOUR_API_KEY' with your News API key
    api_key = 'cf142db758e548d6ab60d5a51e61b208'

    # Replace the list with the stock symbols you're interested in
    symbols = ['WEN']

    fetch_stock_news(api_key, symbols)
```

```python
import pandas as pd
import yfinance as yf
from datetime import datetime

# Define the list of tickers
tickers = [	"BRK-B","PBR-A","BF-A","PDX","HEI-A","UHAL-B","EQH","SSB","MOG-B","MOG-A","GEF-B","LGF-A","AKO-B","AKO-A"]

# Create an empty dictionary to store the ticker-industry mapping
ticker_info = {}

# Loop through each ticker
for ticker in tickers:
    # Create a Ticker object
    ticker_obj = yf.Ticker(ticker)

    # Get the summary profile for the ticker
    summary_profile = ticker_obj.info

    # Extract the industry and full name information from the summary profile
    industry = summary_profile.get('industry', '')
    full_name = summary_profile.get('longName', '')

    # Store the ticker, industry, and full name in the dictionary
    ticker_info[ticker] = {'Industry': industry, 'FullName': full_name}

# Convert the dictionary to a Pandas DataFrame
df_info = pd.DataFrame.from_dict(ticker_info, orient='index')

# Display the DataFrame
display(df_info)

# Export the DataFrame to Excel
excel_file_path = 'stock_info.xlsx'  # Specify the file path
df_info.to_excel(excel_file_path)

print(f'DataFrame exported to Excel file: {excel_file_path}')
```
